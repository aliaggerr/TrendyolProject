```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium. common. exceptions import TimeoutException
from selenium.common.exceptions import TimeoutException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
```

```python
mylist=[]
x=1
while x<=42:
    
    r=requests.get(f"https://www.trendyol.com/sr/kadin-giyim-x-g1-c82?sst=BEST_SELLER&pi={x}")
    soup=BeautifulSoup(r.content,"lxml")
    bir=soup.find_all("div",attrs={"class ","p-card-chldrn-cntnr"})
    time.sleep(0.01)
    for st in bir[:1000]:
        end_link=st.a.get("href")
        head_link="https://www.trendyol.com"
        Link=head_link+end_link
        browser=webdriver.Chrome("C:\Selenium\chromedriver")
        browser.get(Link)
        
        try:
            Saler=browser.find_element("xpath", '//*[@id="product-detail-app"]/div/div[2]/div[1]/div[2]/div[2]/div[2]/div/div/div[2]/a').text
        except:
            Saler=None
        try:
            Brand=browser.find_element("xpath", '//*[@id="product-detail-app"]/div/article[2]/div/div[1]/div/div[1]/h3/span').text
        except:
            Brand=None
        try:
            Name_of_Product=browser.find_element("xpath",'//*[@id="product-detail-app"]/div/div[2]/div[1]/div[2]/div[2]/div[2]/div/div/div[1]/h1/span').text
        except:
            Name_of_Product=None
        try:
            Type_of_Clothes=browser.find_element("xpath", '//*[@id="marketing-product-detail-breadcrumb"]/div/a[4]/span').text
        except:
            Type_of_Clothes=None
        try:
            Count_of_Rating=browser.find_element("xpath",'//*[@id="product-detail-app"]/div/div[2]/div[1]/div[2]/div[2]/div[2]/div/div/div[3]/div/a[1]').text
        except:
            Count_of_Rating=None
        try:
            Score_of_Rating=browser.find_element("xpath","//*[@id='product-detail-app']/div/article[1]/div/div[1]/div/div[1]/span").text
        except:
            Score_of_Rating=None
        try:
            Price=browser.find_element("xpath",'//*[@id="product-detail-app"]/div/div[2]/div[1]/div[2]/div[2]/div[2]/div/div/div[4]/div/div').text
            
        except:
            Price=None
        
        try:
            Score_of_Saler=browser.find_element("xpath","//*[@id='product-detail-app']/div/div[2]/div[2]/section/div[1]/div/div/div[2]").text
        except:
            Score_of_Saler=None
        try:
            Count_of_Question=browser.find_element("xpath",'//*[@id="product-detail-app"]/div/div[2]/div[1]/div[2]/div[2]/div[2]/div/div/div[3]/div/a[2]').text
        except:
            Count_of_Question=None
        
        time.sleep(2)
        try:
            element = browser.find_element_by_css_selector('#product-detail-app > div > div.pr-rnr-w > div.pr-rnr-cn.gnr-cnt-br > div.pr-rnr-com-w > a')
            browser.execute_script("arguments[0].click();", element)
        except:
            pass
     
        mylist.append([Link,Saler,Brand,Name_of_Product,Type_of_Clothes,Count_of_Rating,Score_of_Rating,Price,Score_of_Saler,Count_of_Question])
    browser.quit()
    
    x+=1
    
            
```

```python
df=pd.DataFrame(mylist,columns=["Link","Saler","Brand","Name_of_Product","Type_of_Clothes","Count_of_Rating","Score_of_Rating","Price","Score_of_Saler","Count_of_Question",])
```

```python
df.to_excel("trendyol_bestsaller_data.xlsx")
```
